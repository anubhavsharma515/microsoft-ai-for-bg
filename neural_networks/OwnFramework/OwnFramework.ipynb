## Machine Learning Problem

Suppose we have input dataset $\langle X,Y\rangle$, where $X$ is a set of features, and $Y$ - corresponding labels. For regression problem, $y_i\in\mathbb{R}$, and for classification it is represented by a class number $y_i\in\{0,\dots,n\}$.

Any machine learning model can be represented by function $f_\theta(x)$, where $\theta$ is a set of **parameters**. Our goal is to find such parameters $\theta$ that our model fits the dataset in the best way. The criteria is defined by **loss function** $\mathcal{L}$, and we need to find optimal value
$$
\theta = \mathrm{argmin}_\theta \mathcal{L}(f_\theta(X),Y)
$$
Loss function depends on the problem being solved

### Loss functions for regression
For regression, we often use **abosolute error** $\mathcal{L}_{abs}(\theta) = \sum_{i=1}^n |y_i - f_{\theta}(x_i)|$, or **mean squared error**: $\mathcal{L}_{sq}(\theta) = \sum_{i=1}^n (y_i - f_{\theta}(x_i))^2$

```python
def plot_loss_func(suptitle, functions, ylabels, xlabel):
    fig, ax = plt.subplots(1, len(functions), figsize=(9, 3))
    plt.subplots_adjust(bottom=0.2, wspace=0.4)
    fig.suptitle(suptitle)
    for i, fun in enumerate(functions):
        ax[i].set_xlabel(xlabel)
        if len(ylabels) > i:
            ax[i].set_ylabel(ylabels[i])
        ax[i].plot(x, fun) # x passed as an array of vals b/w -2, 2 below, fun is simply the y-value for corresponding x based on function
```

```python
x = np.linspace(-2, 2, 101)
plot_loss_func(
    suptitle = 'Common loss functions for regression',
    functions = [np.abs(x), np.power(x, 2)],# [0] = absoltue loss, [1] = square loss
    ylabels   = ['(absolute loss)',
                 '(squared loss)'],
    xlabel    = '$y - f(x_i)$'
)
```

### Loss functions for classification,

Let's consider binary classification for a moment. In this case we have two classes, numbered 0 and 1. The output of the network $f_\theta(x_i)\in [0,1]$ essentially defines the probability of choosing the class 1.,

**0-1 loss**

0-1 loss is the same as calculating accuracy of the model - we compute the number of correct classifications:,

$$\mathcal{L}_{0-1} = \sum_{i=1}^n l_i \quad  
        l_i = 
       \begin{cases}
         1 & (f(x_i)<0.5 \land y_i=0) \lor (f(x_i)<0.5 \land y_i=1) \\
         0 & \mathrm{otherwise}
       \end{cases}
$$

This is saying that $l_i$ is 1 (penalty of 1) if either:

* the output from the model ($f(x_i)$) is < 0.5 and if the true label is 0 (incorrect classification)
* the output from the model ($f(x_i)$) is < 0.5 and if the true label is 1 (incorrect classification)



**The threshold of 0.5 here depends entirely on the choice of inputs and labels. This could also be 0 (and the results could be <0 if it's not class=1 or >0 if class=1)**

However, accuracy itself does not show how far are we from the right classification. It could be that we missed the correct class just by a little bit, and that is in a way \"better\" (in a sense that we need to correct weights much less) than missing significantly. Thus, more often logistic loss is used, which takes this into account.
We used 0-1 loss in the previous perceptron - just calculating the total number of correct classifications (both +ve and -ve).

**Logistic Loss**

$$\mathcal{L}_{log} = \sum_{i=1}^n -y\log(f_{\theta}(x_i)) - (1-y)\log(1-f_\theta(x_i))$$

```python
# X is here represents probabilities that the model outputs depending on the class.
x = np.linspace(0, 1, 100)

# Loss is minimum when the probability < 0.5 and the the label is actually correct
def zero_one(d):
    if d < 0.5:
        return 1
    return 0
zero_one_v = np.vectorize(zero_one)

def log_loss(fx):
    # assumes y == 1
    return -np.log(fx)
    # assumes y==0, inverse of the graph where y==1
    # return -np.log(1-fx)
```

```python
plot_loss_func(suptitle = 'Common loss functions for classification (class=1)',
                   functions = [zero_one_v(x), log_loss(x)],
                   ylabels    = ['(0-1 loss)',
                                 '(logistic loss)'],
                   xlabel     = '$p$')
```

# Multi-layer Network example
The model below demonstrates working of a multi-layer perceptron (2 layers to begin with) yielding 2 output layers (each representing
the probability of the input belonging to a particular class)


The main difference between this and the previous single Perceptron example is SIMPLY THE EXISTENCE OF ONLY 1 OUTPUT LAYER (HENCE THE WEIGHT x FEATURE GIVES us a 2x1 entry for each input (each entry in this array corresponds to how likely something is to a particular class)).

```python
class Linear:
    def __init__(self,nin,nout):
        self.W = np.random.normal(0, 1.0/np.sqrt(nin), (nout, nin))
        self.b = np.zeros((1,nout))
        
    def forward(self, x):
        return np.dot(x, self.W.T) + self.b
    
net = Linear(2,2)
foo = net.forward(train_x[0:5])
```

# Softmax
This is a function used to normalize a set of values in such a way that the sum of all values normalized is returned to be 1.

The function looks like:

$$
  \sigma(z_c) = \frac{e^{z_c}}{\sum_{j} e^{z_j}}, \quad\mathrm{for}\quad c\in 1 .. |C|
$$

So in the output above of w.x + b, the first result (with $res[0] = Y(class_1)$ and $res[0] = Y(class_2)$) of applying softmax will be on each column of a particular row (so each output for a given input example).


```python
class Softmax:
    def forward(self,z):
        zmax = z.max(axis=1,keepdims=True) # For each row, return the column with max value, if shape is 5,2 return 5,1 instead of 1,5
        expz = np.exp(z-zmax)
        Z = expz.sum(axis=1,keepdims=True)
        return expz / Z

softmax = Softmax()
softmax.forward(net.forward(train_x[0:10]))
```

# Cross Entropy Loss

A loss function in classification is typically a logistic function, which can be generalized as cross-entropy loss. Cross-entropy loss is a function that can calculate similarity between two arbitrary probability distributions. You can find more detailed discussion about it on Wikipedia.

**LOG LOSS IS JUST BINARY. CROSS ENTROPY IS GENERALIZATION OF LOG LOSS TO MULTI-CLASS CLASSIFICATION**

```python
def plot_cross_ent():
    p = np.linspace(0.01, 0.99, 101) # estimated probability p(y|x)
    cross_ent_v = np.vectorize(cross_ent) # So cool - it's vectorizing the return value of a func, but the func needs to be passed
                                          # and returned first
    f3, ax = plt.subplots(1,1, figsize=(8, 3))
    l1, = plt.plot(p, cross_ent_v(p, 1), 'r--') # Gets loss for true class = 1
    l2, = plt.plot(p, cross_ent_v(p, 0), 'r-')  # Gets loss for true class = 0
    plt.legend([l1, l2], ['$y = 1$', '$y = 0$'], loc = 'upper center', ncol = 2)
    plt.xlabel('$\hat{p}(y|x)$', size=18)
    plt.ylabel('$\mathcal{L}_{CE}$', size=18)
    plt.show()
```

```python
def cross_ent(prediction, ground_truth):
    t = 1 if ground_truth > 0.5 else 0
    return -t * np.log(prediction) - (1 - t) * np.log(1 - prediction) # Same as log loss function defined in graph in cell 9
plot_cross_ent()
# p(y/x) = probabity of seeing class y = 1, given the input x.
```
